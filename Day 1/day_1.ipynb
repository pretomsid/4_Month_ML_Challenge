1. Micro Challenge : The Identity Swap 

The core learning is that in Python, you can swap two variable values in a single line (e.g., x, y = y, x) by using Tuple Unpacking. This trick works because Python secretly creates a temporary data group (a Tuple) to hold the values, which are then instantly reassigned.

```python
x = 100
y = 200
x,y=y,x
print(x,y)
```

2. Micro Challenge : The Type Editor

This drill proves that the input() function is strictly a string-fetcher. The "transformation" in Step 2 (casting to a float) does not change the original object because strings are immutable. Instead, the float() constructor allocates a brand new object at a different memory address.

```python
number = input("Enter a number: ")
print(type(number))

number = float(number) # Convert string to float (typecasting)
print(type(number))
```

3. Micro Challenge : The Precision Banker

The learning from the Precision Banker micro-challenge is about how Python handles floating-point numbers and the role of F-Strings in formatting.

Floating-Point Representation: Python floats use IEEE 754 double-precision (64-bit). The actual number stored in memory may have a very high precision (e.g., 33.333333333333336 for 100/3).

F-Strings as a View Layer: F-Strings act as a "View Layer". They mask the raw precision for the user interface without changing the underlying data integrity in memory.

```python
bill = 100 / 3
print(round(bill,3)) # Rounding to 3 decimal places, round(variable, no_of_decimal_places)
```

4. Micro Challenge : The Modulo Architect

The learning is about using Integer Division (`//`) and the Modulo operator (`%`) to solve problems that involve cycles, like converting seconds into hours, minutes, and seconds.

The mechanics taught are:

  * **Integer Division (`//`):** Used to calculate the number of full units (e.g., Hours: `total // 3600`) by discarding the remainder.
  * **Modulo Operator (`%`):** Used to find the remaining amount after division (e.g., Remaining Seconds: `total % 3600`), keeping only the remainder.

This logic is foundational for cyclic algorithms used in fields like cryptography, hash tables, and time series processing.


```python
total_seconds = int(input())

Hours = total_seconds // 3600 
Minutes = (total_seconds % 3600)// 60
Seconds = total_seconds % 60 

print(Hours,Minutes,Seconds)
print(f"{Hours}:{Minutes}:{Seconds}")
```
